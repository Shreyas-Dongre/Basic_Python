#Task 1:  
Write a Python program to display all the prime numbers between any two numbers taken from the user


```
a=int(input('Enter lower bound of the range.'))
b=int(input('Enter upper bound of the range.'))
for i in range(a,b+1):
  if i>1:
    for j in range(2,i):
      if(i%j)==0:
        break
    else:
      print(i)    
```

#Task 2:  
Write a program to check whether a user-entered number is a Perfect Number.


```
a=int(input("Enter a number which you have to check. "))
b=0
for i in range(1,a):
  if(a%i==0):
    b+=i
if(b==a):
  print("The number is a perfect number :)")
else:
  print("The number is NOT a perfect number.")      
```

#Task 4:

                        string2: Mary Likes Python

```
a=str(input('Enter string 1 : '))
b=str(input('Enter string 2 : '))
c=[]
d=[]
for char in a :
  if char.isupper():
    c.append(char)
  #print(c)
    
for char in b :
  if char.isupper():
    d.append(char)
  #print(d)
print(c)
print(d)  

res=c+d
print(res)  
```

#Task 10:
Write a Python program to construct the following pattern {^ is used in place of *}

       ^                          
      ^^^                      
     ^^^^^                     
      ^^^                   
       ^  


```
a = 5

for i in range(1, (a+1)//2 + 1): 
    for j in range((a+1)//2 - i):
        print(" ", end = "")
    for k in range((i*2)-1):
        print("*", end = "")
    print()

for i in range((a+1)//2 + 1, a + 1): 
    for j in range(i - (a+1)//2):
        print(" ", end = "")
    for j in range((a+1 - i)*2 - 1):
        print("*", end = "")
    print()
```

#Task 11: 
Write a program to take integers from the user in a list. Removes elements of the list that are the same and retain. 
Input: ([1,1,2,2])
Expected Output : [1,2].


```
l=[]
a=int(input('Enter the number of list elements '))
for i in range(0,a):
  b=int(input('Enter elements '))
  l.append(b)
l1=[]
for i in l:
  if i not in l1:
    l1.append(i)
print('Your list',l)
print('After removing Duplicates',l1)      
```
