#Task 1
Write a function that returns the sum of multiples of 3 and 5 between 0 and limit (parameter).            
Test case : limit = 20, it should return (sum of 3, 5, 6, 9, 10, 12, 15, 18, 20)


```python
def lim(limit):
  s=0
  for i in range(0,limit+1):
    if i%3==0 or i%5 ==0 :
      s=s+i
     
  return s    
  print('The sum of multiples is ',s)    
  
limit=int(input('Enter the limit parameter '))
lim(limit)  
  
```

#Task 2  
Write a Python function to check whether a string is a pangram or not.


```python
import string

letters=set(string.ascii_lowercase)
def ispan(a):
  res=set(a.lower()) >= letters
  return res

a=input('Enter the string ')
if (ispan(a) == True):
  print('Given string is a Pangram.')
else:
  print('Given string is NOT a Pangram.')    
```

#Task 3
Write a function called show numbers that takes a parameter called limit.     
 It should print all the numbers between 0 and limit with a label to identify the even and odd numbers.                  
For example, if the limit is 4, it should print:                 
0 EVEN             
1 ODD                 
2 EVEN                
3 ODD              
4 EVEN      


```python
def show_numbers(limit):
  for i in range (0,limit+1):
    if i%2==0:
      print(i,"EVEN")
    else:
      print(i,"ODD")
  return

a=int(input('Enter a number '))
show_numbers(a)        
```

Maintain the Speed!”.                                                                                Otherwise, for every 5 units above the speed limit (70), it should give the driver one penalty point and print the total number of penalty points.
```python
def check(speed):
  points = 70
  point_value = speed - points
  point = point_value / 5
  if speed == 70:
    print("Good! Maintain the speed!")
  elif point>=12:
    print("Penalty Points",point)
    print('License suspended')
  else:
    print("Penalty Points",point)

a=int(input('Enter value of speed '))
check(a)
   
```

#Task 6 
Write a function to check whether two strings are anagrams of each other.          
Test Cases :            
Stressed - Desserts → True                   
Inch - Chin → True               
Inches - Chin → False


```python
def check(s1,s2):
  if(sorted(s1) == sorted(s2)):
    print('The strings are anagrams. TRUE')
  else :
    print('The strings are not anagrams. FALSE')

n=3
while n>0:
  s1=input('Enter first string ')    
  s2=input('Enter second string ')  
  check(s1,s2)
  n=n-1
```

#Task 7
Write a function to detect an anagram from the string. (Anagram Scramble) (i.e a particular word can be framed or not from another.)           
Test case :

```python
def anagram(s1,s2):
  for i in s1.lower():
    if i in s2.lower():
      return True
    return False
n = input("enter string:")
res = anagram(n,"laboratory")
if res:
  print("Anagram")
else:
  print("Not an anagram")
```

#Task 10             
Write a Function to rotate string left and right by x length         
Input : string =  “Bengaluru”, x =3               
Output:         
Left Rotation :  galuruBen           
Right Rotation :  uruBengal


```python
def rotate(s,x):
  l1= s[0:x]
  l2= s[x:]
  r1=s[0:len(s)-x]
  r2=s[len(s)-x:]

  print('Left Rotation: ',(l2+l1))
  print('Right Rotation: ',(r2+r1))
  return

a= input('Enter a string ')
b= int(input('Enter the length of rotation '))  
rotate(a,b)
```
