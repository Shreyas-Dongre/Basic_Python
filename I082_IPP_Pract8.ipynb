#3.1
Write a function average that takes a list of numbers and returns the average.                          
1.	Define a function called average that has one argument, numbers.
2.	Inside that function, call the built-in sum() function with the numbers list as a parameter.                                                           Store the result in a variable called total.
3.	Use float() to convert total and store the result in total.
4.	Divide total by the length of the numbers list. Use the built-in len() function to calculate that.Return that result.


```python
def average(numbers):
  total = sum(numbers)
  total = float(total)
  avg = total / (len(numbers))
  
  return avg

```


#3.2
Write a function called get_average that takes a student dictionary (like  lloyd, alice, and ritchie) as input and returns his/her weighted average.


```python
def get_average(student):
  homework = average(student["homework"])
  quizzes = average(student["quizzes"])
  tests = average(student["tests"])
  final = homework + quizzes + tests
  
  return final
  
```

#Task 8
Write a Python Program to Create a Dictionary with Key as First Character and Value as Words Starting with that Character.                        
Example:                    
Input:                
string: i love grapes guava watermelon pineapple pear mango            
Output:               
{'i': ['i'], 'l': ['love'], 'g': ['grapes', 'guava'], 'w': ['watermelon'], 'p': ['pineapple', 'pear'], 'm': ['mango']}


```python
string = 'i love grapes guava watermelon pineapple pear mango'

s1 = string.split()
dict1 = {}

for i in s1:
  if (i[0] not in dict1.keys()):
    dict1[i[0]] = []
    dict1[i[0]].append(i)
  else:
    if (i not in dict1[i[0]]):
      dict1[i[0]].append(i)
print(dict1)        
```

#Task 9
List of participants in events A and B            
A → John, Olive,Alfie,Oscar,Ella,William,Harry.            
B → Edward, Julia, Robert, Olive,Alfie,Ella.            
1.	Find out which members attended both events.
2.	Find out which members attended only one of the events. **OMIT**
3.	Find out which members attended only one event and not the other.
4.	Receive a list of all participants


```python
A = {'John', 'Olive', 'Alfie', 'Oscar', 'Ella', 'William', 'Harry'}
B = {'Edward', 'Julia', 'Robert', 'Olive', 'Alfie', 'Ella'}

#1
res = A.intersection(B)
print('Members who attended both the events are:',res)


#2 -  OMITTED


#3
res3 = A ^ B
print('Members who  attended ONLY one of the events are:',res3)


#4
res4 = A.union(B)
print('List of all participants:',list(res4))
```

```python
S1 = {'purple', 'red', 'blue', 'black','pink','yellow','orange','white'}
S2 = {'yellow', 'orange', 'black','purple','pink','brown'}
S3 = {'pink', 'purple','orange'}
```

```python
#10.1
print('S1 =',S1)
print('S2 =',S2)
print('S3 =',S3)
print('\n')

print('Is S1 subset of S2?')
print(S1.issubset(S2))

print('Is S2 subset of S1?')
print(S2.issubset(S1))

print('Is S1 subset of S3?')
print(S1.issubset(S3))

print('Is S3 subset of S1?')
print(S3.issubset(S1))

print('Is S2 subset of S3?')
print(S2.issubset(S3))

print('Is S3 subset of S2?')
print(S3.issubset(S2))
```

```python
#10.2
print('S1 =',S1)
print('S2 =',S2)
print('S3 =',S3)
print('\n')

print('Is S1 superset of S2?')
print(S1.issuperset(S2))

print('Is S2 superset of S1?')
print(S2.issuperset(S1))

print('Is S1 superset of S3?')
print(S1.issuperset(S3))

print('Is S3 superset of S1?')
print(S3.issuperset(S1))

print('Is S2 superset of S3?')
print(S2.issuperset(S3))

print('Is S3 superset of S2?')
print(S3.issuperset(S2))
```
