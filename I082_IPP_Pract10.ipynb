```python
#Task 1 Solution
class Error(Exception): pass
class NegaSqrtError(Error): pass
def root(a,b,c):
  l1=[a,b,c]
  res  = a ** 0.5
  res2 = b ** 0.5
  res3 = c ** 0.5
  total = res+res2+res3
  
  for i in l1:
    if i<0:
      raise NegaSqrtError
    elif i == 0:
      raise Error
    elif type(i) != float:
      raise Error
    
        
  print('The addition of square roots of numbers',a,b,c,'is:',total)            
a = float(input('Enter 1st number: '))
b = float(input('Enter 2nd number: '))
c = float(input('Enter 3rd number: ')) 
root(a,b,c)   
```

○	If the second input is not '+' or '-', again raise a FormulaError

calculate(2,"*", 5) → 10 
calculate(10,"-", 5) →5 


```python
class Error(Exception):pass
class FormulaError(Exception): pass
while True:
  a = input('Enter 1st number: ')
  b = input('Enter Operator strictly (+,-,*,/) ONLY: ')
  c = input('Enter 2nd number: ')
  d = a+b+c

  if a.lower()=='quit':
    print('ThankYou')
    break
  
  chk=d.split()  
  l = [a,b,c]
  try:
    if len(l) != 3 :
      raise FormulaError('Input does not consist of 3 elements.')
  
    a = float(a)
    c = float(c)  
  

    if b not in ['+','-','*','/']:
      raise FormulaError('{0} is not a valid. Choose an operator from the given range'.format(b))
    if b == '+':
      add = a+c
      print(a,'+',c,'➞',add)
    elif b == '-':
      sub = a-c
      print(a,'-',c,'➞',sub)
    elif b == '*':
      mul = a*c
      print(a,'*',c,'➞',mul)
    elif b == '/':
      if b == 0:
        raise FormulaError
      div = a/c
      print(a,'/',c,'➞',div)
  except ValueError:
        raise FormulaError
  except FormulaError:
        print('Wrong Formula. Please try again!')      
    


```

#Task 4  
Given an integer, create a function that returns the next prime.                  
If the number is prime, return the number itself.        
Examples :     
prime(12) ➞ 13        
prime(24) ➞ 29         
prime(11) ➞ 11


```python
def prime(num):
  k=0
  for i in range(2,num):
      if num%i==0: 
        return prime(num+1)
      else: 
        k=0
  if k==0:
    return num
 
print("prime(24) ➞",prime(24))   
```
