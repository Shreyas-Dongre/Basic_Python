#Task 4
WAP to print duplicates from a list of integers.

```
l=[]
a=int(input('Enter no of elements to be included in list'))
for i in range(0,a):
  b=int(input('Enter the numbers '))
  l.append(b)
  i+=1
print('The list is: ',l)
leng=len(l)
for i in range(0,leng):
  for j in range(i+1,leng):
    if (l[i]==l[j]):
      print('Duplicates are: ',l[j])
```

#Task 5
WAP to check if a string is palindrome or not. (with and without slicing)

```
# With Slicing
def isPalindrome(s):
    return s == s[::-1]
 
s =input('Enter a string ')
ans = isPalindrome(s)
 
if ans:
    print("Yes")
else:
    print("No")
```

```
# Without Slicing
n = input("ENTER  STRING\n")
length = int(len(n))

for i in range(length):
   if n[i] == n[length-i-1]:
     print('Given string is a Palindrome.')
     break
   else:
      print('Given string is NOT a Palindrome.')
      break
```

#Task 10
Write a Python program to check the validity of password input by users.       
Validation:             
1.	At least 1 letter between [a-z] and 1 letter between [A-Z].
2.	At least 1 number between [0-9].
3.	At least 1 character from [$#@].
4.	Minimum length 6 characters.
5.	Maximum length of 16 characters.


```
import re
password=input('Enter your pass to check ')
flag=1
while flag:
  if not re.search('[a-zA-Z]',password):
    break
  if not re.search('[0-9]',password):
    break
  if (len(password)<6 or len(password)>16):
    break
  if not re.search('[$#@]',password):
    break
  else:
    flag=0
    print('Valid')
    break

if flag:
  print('Invalid')            
```
