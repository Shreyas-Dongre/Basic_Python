#Task 7 :
Write a program to check the user entered number is an Armstrong Number

```
num = int(input("Enter a number: "))
s = 0
temp = num
while temp > 0:
   digit = temp % 10
   s = s+digit ** 3
   temp = temp//10
if num == s:
   print(num,"is an Armstrong number")
else:
   print(num,"is not an Armstrong number")
```
