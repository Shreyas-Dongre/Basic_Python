## Basic plotting

```
import matplotlib.pyplot as plt
import numpy as np
```

```
from math import pi
x = np.linspace(-pi, pi, 200)
y = np.sin(2*x)

plt.plot(x, y)
plt.grid()
plt.show()
```

```
from math import pi
x = np.linspace(-pi, pi, 200)
y = np.sin(2*x)
z = np.cos(2*x)

plt.plot(x, y)
plt.plot(x,z,'k--')
plt.legend(['sin','cos'])
plt.grid()
plt.show()
```

```
from math import pi
x = np.linspace(-pi, pi, 200)
y = np.sin(4*x)
z = np.cos(4*x)

plt.plot(x, y, color='k', linestyle='--', linewidth=3, marker='.')
plt.plot(x, z, color='r', linestyle='-', marker='o')
plt.legend(['sin','cos'])
plt.grid()
plt.show()
```

```
y = np.exp(x)

plt.plot(x, y)
plt.grid()
plt.show()
plt.savefig('exp.png')
```

# What is a histogram?
* It is a graphical representation of data using
bars of different heights
* Histogram groups numbers into ranges and
the height of each bar depicts the frequency
of each range or bin
## When to use histograms?
* To represent the frequency distribution of
numerical variables

```
# Histogram of Age
x = [1,1,2,3,3,5,7,8,9,10,10,11,11,13,13,15,16,17,18,18,18,19,20,21,21,23,24,
     24,25,25,25,25,26,26,26,27,27,27,27,27,29,30,30,31,33,34,34,34,35,36,
     36,37,37,38,38,39,40,41,41,42,43,44,45,45,46,47,48,48,49,50,51,52,53,54,
     55,55,56,57,58,60,61,63,64,65,66,68,70,71,72,74,75,77,81,83,84,87,89,
     90,90,91]

plt.hist(x, bins=10, rwidth = 0.9) # rwidth: bar's relative width, default 1
plt.grid(axis='y',alpha=0.5) # Only the y grid
plt.xlabel('Age value')
plt.ylabel('Frequency')
plt.title('My first histogram')
plt.show()
```

```
plt.hist(x, bins=10, color = 'yellow', edgecolor ='black', rwidth = 0.9) 

plt.grid(axis='y',alpha=0.5) # Only the y grid
plt.xlabel('Age value')
plt.ylabel('Frequency')
plt.title('My first histogram')
plt.show()
```

# What is a bar plot?
* A bar plot is a plot that presents categorical
data with rectangular bars with lengths
proportional to the counts that they
represent
# When to use bar plot?
* To represent the frequency distribution of
categorical variables
* A bar diagram makes it easy to compare sets
of data between different groups

```
counts = [60, 20, 35]
programme = ('BTechIntegrated', 'MCA', 'BTech')

plt.bar(programme, counts)
plt.xlabel('Programme Type')
plt.ylabel('Count')
plt.title('Distribution')
plt.show()
```

```
import numpy as np
counts = [60, 20, 35]
programme = ('BTechIntegrated', 'MCA', 'BTech')
index= np.arange(len(programme))

plt.bar(index, counts, color = ['red','blue','cyan'])
plt.title('Bar blot of student distribution for workshop')
plt.xlabel('Programme Type')
plt.ylabel('Count')
plt.xticks(index,programme,rotation =90)
plt.show()
```

# What is Pie chart
* A Pie Chart displays data in a circular graph. 
* The pieces of the graph are proportional to the fraction of the whole in each category, each slice of the pie is relative to the size of that category in the group as a whole.
* The entire pie represents 100 percent of a whole, while the pie slices represent portions of the whole.
# When to use Pie chart
* Pie charts are good for displaying data for around 6 categories or fewer.

```
# Pie chart
slices = [59291, 40000, 47544, 35000, 35917]
labels = ['Javascript', 'HTML/CSS', 'Python', 'SQL', 'Java']

plt.pie(slices, labels=labels)
plt.title('My Pie chart')
plt.show()
```

```
# Pie chart
slices = [59291, 40000, 47544, 35000, 35917]
labels = ['Javascript', 'HTML/CSS', 'Python', 'SQL', 'Java']
explode = [0,0,0.1,0,0]
color=['r','b','g','c','y']

plt.pie(slices,labels=labels,explode=explode,startangle=90,autopct='%1.2f%%',colors=color)
plt.title('My Pie chart')
plt.show()
```

# What is a scatter plot?
* A scatter plot is a set of points that represents
the values obtained for two different variables
plotted on a horizontal and vertical axes
# When to use scatter plots?
* Scatter plots are used to convey the relationship
between two numerical variables
* Scatter plots are sometimes called correlation
plots because they show how two variables are
correlated

```
x = [5,7,8,7,2,17,2,9,4,11,12,9,6] # age of car
y = [99,86,87,88,111,86,103,87,94,78,77,85,86] # speed of car

plt.scatter(x, y, c='red')
plt.title('My scatter plot')
plt.xlabel('Age')
plt.ylabel('Speed')
plt.show()
```

# Boxplot

A box and whisker plot—also called a box plot—displays the five-number summary of a set of data. The five-number summary is the minimum, first quartile, median, third quartile, and maximum.
A boxplot is a graph that gives you a good indication of how the values in the data are spread out.

## BoxPlot

Five-number summary:

1.   min = minimum value
2.   25% = first quartile (Q1) = median of the lower half of the data
3.   50% = second quartile (Q2) = median of the data
4.   75% = third quartile (Q3) = median of the upper half of the data
5.   max = maximum value
