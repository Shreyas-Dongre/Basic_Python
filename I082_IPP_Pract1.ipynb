##Task 2:  
Write a Python Program to the roots of a Quadratic Equation. d = b2 - 4ac


```
import cmath
a=int(input('Enter a '))
b=int(input('Enter b '))
c=int(input('Enter c '))

d=(b**2)-4*(a)*(c)
root1=(-b+cmath.sqrt(d))/(2*a)
root2=(-b-cmath.sqrt(d))/(2*a)

print('Roots of Quadratic Equation are {} and {}'.format(root1,root2))
```

##Task 7 :
1.	Write a Python program to convert seconds to day, hour, minutes and seconds.
2.	Write a Python program to convert the number of days to years, months & days.


```
#1
seconds=float(input('Enter number of seconds'))
day=int(seconds/86400)
a=(seconds%86400)
hours=int(a/3600)
b=(a%3600)
minutes=int(b/60)
seconds=(b%60)
print('Days:', day)
print('Hours:',hours)
print('Minutes',minutes)
print('seconds',seconds)
```

```
#2
days=float(input('Enter no of days  '))
years= int(days/365)
a=(days%365)
months=int(a/30)
b=(a%30)
days=int(b)

print('No of years',years)
print('No of months',months)
print('No of days',days)
```

##Task 8 :

```
#1
import math
a=int(input('Enter len of major axis '))
b=int(input('Enter len of minor axis '))
area=math.pi*a*b
print("Area of ellipse ",area)
```
